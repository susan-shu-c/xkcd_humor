## xkcd

This humorous plot was a fun way to learn the functionalities of matplotlib, in the style of webcomic xkcd.

```python
from matplotlib import pyplot as plt
import numpy as np

plt.xkcd()

weeks = np.arange(14)
ditems = [1, 0, 2, 1, 3,
         1, 1, 2, 1,
         0, 1, 10, 0]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.xticks([0, 5, 9], ('Feb', 'Mar', 'Apr'))
plt.yticks([])

plt.ylim(0,10)

plt.annotate(
    'final exams and papers\n adjusted suffer value',
    xy=(11, 10), arrowprops=dict(arrowstyle='->'), xytext=(5,7), fontsize=12)

plt.plot(ditems)

plt.xlabel('time')
plt.ylabel('there\'s a reason they\'re called DEADlines\n- Dr. Trien Nguyen')
plt.title('frequency of suffering', fontsize=19)
plt.show()

plt.savefig('suffering.png')
```
